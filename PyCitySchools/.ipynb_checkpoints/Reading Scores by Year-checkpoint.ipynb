```python
# Dependencies
import pandas as pd
```

```python
# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()    
```

```python
#Reading Score for Year 9

reading_year9_df = school_data_complete_df.loc[school_data_complete_df["year"] == 9,["school_name","reading_score"]]
#reading_year9_df 

readingyear9_group = reading_year9_df.groupby(["school_name"])
avg_reading_year9 = readingyear9_group.mean()
avg_reading_year9 = avg_reading_year9.rename(columns={"reading_score":"Year 9"})
avg_reading_year9.head()
```

```python
#Reading Score for Year 10

reading_year10_df = school_data_complete_df.loc[school_data_complete_df["year"] == 10,["school_name","reading_score"]]
#reading_year10_df 

readingyear10_group = reading_year10_df.groupby(["school_name"])
avg_reading_year10 = readingyear10_group.mean()
avg_reading_year10 = avg_reading_year10.rename(columns={"reading_score":"Year 10"})
avg_reading_year10.head()
```

```python
#Reading Score for Year 11

reading_year11_df = school_data_complete_df.loc[school_data_complete_df["year"] == 11,["school_name","reading_score"]]
#reading_year11_df 

readingyear11_group = reading_year11_df.groupby(["school_name"])
avg_reading_year11 = readingyear11_group.mean()
avg_reading_year11 = avg_reading_year11.rename(columns={"reading_score":"Year 11"})
avg_reading_year11.head()
```

```python
#Reading Score for Year 12

reading_year12_df = school_data_complete_df.loc[school_data_complete_df["year"] == 12,["school_name","reading_score"]]
#reading_year11_df 

readingyear12_group = reading_year12_df.groupby(["school_name"])
avg_reading_year12 = readingyear12_group.mean()
avg_reading_year12 = avg_reading_year12.rename(columns={"reading_score":"Year 12"})
avg_reading_year12.head()
```

```python
#Combine all Reading scores by year
reading_scores_by_year = pd.merge(pd.merge(pd.merge(avg_reading_year9, avg_reading_year10, on="school_name"),
                                       avg_reading_year11, on="school_name"),avg_reading_year12, on="school_name")
reading_scores_by_year
```
