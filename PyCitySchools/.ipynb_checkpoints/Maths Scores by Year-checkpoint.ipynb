```python
# Dependencies
import pandas as pd
```

```python
# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()    
```

```python
#Maths Score for Year 9

maths_year9_df = school_data_complete_df.loc[school_data_complete_df["year"] == 9,["school_name","maths_score"]]
#maths_year9_df 

mathsyear9_group = maths_year9_df.groupby(["school_name"])
avg_maths_year9 = mathsyear9_group.mean()
avg_maths_year9 = avg_maths_year9.rename(columns={"maths_score":"Year 9"})
avg_maths_year9.head()
```

```python
#Maths Score for Year 10

maths_year10_df = school_data_complete_df.loc[school_data_complete_df["year"] == 10,["school_name","maths_score"]]
#maths_year10_df 

mathsyear10_group = maths_year10_df.groupby(["school_name"])
avg_maths_year10 = mathsyear10_group.mean()
avg_maths_year10 = avg_maths_year10.rename(columns={"maths_score":"Year 10"})
avg_maths_year10.head()
```

```python
#Maths Score for Year 11

maths_year11_df = school_data_complete_df.loc[school_data_complete_df["year"] == 11,["school_name","maths_score"]]
#maths_year11_df 

mathsyear11_group = maths_year11_df.groupby(["school_name"])
avg_maths_year11 = mathsyear11_group.mean()
avg_maths_year11 = avg_maths_year11.rename(columns={"maths_score":"Year 11"})
avg_maths_year11.head()
```

```python
#Maths Score for Year 12

maths_year12_df = school_data_complete_df.loc[school_data_complete_df["year"] == 12,["school_name","maths_score"]]
#maths_year11_df 

mathsyear12_group = maths_year12_df.groupby(["school_name"])
avg_maths_year12 = mathsyear12_group.mean()
avg_maths_year12 = avg_maths_year12.rename(columns={"maths_score":"Year 12"})
avg_maths_year12.head()
```

```python
#Combine all maths scores by year
maths_scores_by_year = pd.merge(pd.merge(pd.merge(avg_maths_year9, avg_maths_year10, on="school_name"),
                                       avg_maths_year11, on="school_name"),avg_maths_year12, on="school_name")
maths_scores_by_year
```
