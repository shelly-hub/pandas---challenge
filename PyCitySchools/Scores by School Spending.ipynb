```python
# Dependencies
import pandas as pd
```

```python
# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()    
```

```python
#Group by Schools
school_group = school_data_complete_df.groupby(["school_name","type"])
#school_df = school_group.count()

student_df = pd.DataFrame(school_group[["Student ID"]].count())
student_df.head()
```

```python
#Total  Students is equal to total count of student ID
#Hence needs to rename 
student_df = student_df.rename(columns = {"Student ID":"Total Students"})
student_df.head()
```

```python
#Total School Budget
total_budget_df= pd.DataFrame(school_group["budget"].mean())
total_budget_df.head()
```

```python
#Calculating Per Student Budget
#First combine total school budget and total students
merge_budget_df = pd.merge(student_df,total_budget_df, on=["school_name","type"] )
merge_budget_df.head()
```

```python
#Creating new column per student budget 
merge_budget_df["Per Student Budget"] = (merge_budget_df["budget"]/merge_budget_df["Total Students"])

merge_budget_df= merge_budget_df.rename(columns = {"budget":"Total School Budget"})

merge_budget_df.head()
```

```python
#CALCULATE Average Maths and Reading Score

avg_maths_reading_df = pd.DataFrame(school_group[["maths_score", "reading_score"]].mean())
avg_maths_reading_df = avg_maths_reading_df.rename(columns={"maths_score":"Average Maths Score", 
                                                           "reading_score": "Average Reading Score"})

avg_maths_reading_df.head()
```

```python
#Filter Percentage Maths and Reading Score (50 or greater)
pass_maths_df =  school_data_complete_df.loc[school_data_complete_df["maths_score"] >= 50,:]

maths_grouping = pass_maths_df.groupby(["school_name","type"])
maths_group_df = pd.DataFrame(maths_grouping [["maths_score"]].count())
#maths_group_df

maths_group_df["% Passing Maths"] = (maths_group_df["maths_score"]/student_df["Total Students"])*100

del maths_group_df["maths_score"]

maths_group_df.head()
```

```python
#Filter Percentage Reading Score (50 or greater)
pass_reading_df =  school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 50,:]

reading_grouping = pass_reading_df.groupby(["school_name","type"])
reading_group_df = pd.DataFrame(reading_grouping [["reading_score"]].count())
#reading_group_df

reading_group_df["% Passing Reading"] = (reading_group_df["reading_score"]/student_df["Total Students"])*100

del reading_group_df["reading_score"]

reading_group_df.head()
```

```python
#To calculate overall passing
#Merge Maths & Reading passing percentages

pass_df =  school_data_complete_df.loc[(school_data_complete_df["maths_score"] >= 50) & 
                                          (school_data_complete_df["reading_score"] >= 50),:]
overall_group = pass_df.groupby(["school_name","type"])
overall_group_df = pd.DataFrame(overall_group[["maths_score","reading_score"]].count())
#overall_group_df

overall_group_df["% Overall Passing"] = (overall_group_df["maths_score"]/student_df["Total Students"])*100 

del overall_group_df ["maths_score"]
del overall_group_df ["reading_score"]

overall_group_df.head()
```

```python
#Combine percentages column together
percent_summary_df =maths_group_df.merge(reading_group_df, on =["school_name","type"]).join(overall_group_df, on=["school_name","type"])
percent_summary_df.head()
```

```python
#Combine all dataframe together
school_overall_df = merge_budget_df.merge(avg_maths_reading_df, on = ["school_name","type"]).join(percent_summary_df,on = ["school_name","type"])
school_overall_df.head()
```

```python

#Identify max and min value of per student budget
print(school_overall_df["Per Student Budget"].max())
print(school_overall_df["Per Student Budget"].min())
```

```python
#Create Bins
bins = [0, 585, 630, 645, 680]
budget_range = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_overall_budget_df = school_overall_df
school_overall_budget_df ["Per Student Budget"]= pd.cut(school_overall_df["Per Student Budget"], bins, labels=budget_range)
school_overall_budget_df.head()
```

```python
#Rename Budget
school_overall_budget_df = school_overall_budget_df.rename(columns={"Per Student Budget":"Spending Ranges(Per Student)"})
school_overall_budget_df.head()
```

```python
#Create new Groupby Per budget students
budget_group = school_overall_budget_df.groupby(["Spending Ranges(Per Student)"])

spending_summary = budget_group[["Average Maths Score","Average Reading Score",
                                 "% Passing Maths","% Passing Reading","% Overall Passing"]].mean()

spending_summary
```

```python
#Cleaner Format for Scores by School Spending
spending_summary = spending_summary.style.format("{:,.2f}")
spending_summary
```
